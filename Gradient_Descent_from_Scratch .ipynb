## gradient descent 

```python
import numpy as np 
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.01 
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        
        mse_values=[]
        for j in (y-y_predicted):
            mse_values.append(j**2)
            
        cost = (1/n)*sum(mse_values)#finding Mean squared error 
        
        md=-(2/n)*sum(x*(y-y_predicted)) #partial derivate of m
        bd=-(2/n)*sum(y-y_predicted) #partial derivate of b
        
        m_curr= m_curr-learning_rate*md # to take new step(m = m-learningrate * partial derivative of m)
        b_curr = b_curr-learning_rate*bd #  to take new step(m = m-learningrate * partial derivative of b)
        
        print('slope: ', m_curr,'intercept: ',b_curr, 'cost: ',cost, 'Iterations: ', i)
        plt.plot(x,y_predicted)
```

```python
x= np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```
