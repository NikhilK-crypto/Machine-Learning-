```python
#machine learning using basic formula(y=mx+b)
#sentdex part 7 (check video for formula)
from statistics import mean 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5,6], dtype=np.float64)
y=np.array([5,4,6,5,6,7],dtype=np.float64)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#formula of Slop (m)= {[mean(x) X mean(y)-mean(xy)]/[(mean(x))^2-mean(x^2)]}
#formula of y-intercept(b)= {mean(y) - [m X mean(x)]}
def BestFitSlope_and_intercept(xs,ys): 
    a=(mean(xs)*mean(ys))-(mean(xs*ys))
    b=((mean(xs))*mean(xs))-(mean(xs**2))
    m=a/b
    
    c= mean(ys)-[m*mean(x)]
    return m,c
```

```python
m,b= BestFitSlope_and_intercept(x,y)
```

```python
regression_line=[]
for xi in x:
    regression_line.append((m*xi)+b)    
```

```python
#predicted y values
regression_line=np.array(regression_line)
regression_line=regression_line.reshape(1,6) # to reshape array from (6,1) to (1,6)
regression_line=regression_line.ravel() # to make it 1d array  
regression_line
```

```python
#plotting x values with predicted y values
plt.style.use('fivethirtyeight')
plt.scatter(x,y, color='blue')
plt.plot(x,regression_line, color='skyblue')
plt.show()
```

```python
#taking x value from user to predict its y value

predict_x=float(input('Enter the x value'))
predict_y= m*predict_x + b 
print(predict_y)

plt.style.use('fivethirtyeight')
plt.figure(figsize=(5,4), dpi= 150)

plt.scatter(x,y, color='blue')
plt.scatter(predict_x,predict_y,color='red')
plt.plot(x,regression_line, color='skyblue')
plt.show()
```

```python
#Using R Squared to determine the goodness of fit
# r^2 =1-[SSE]/[SST]
#SSE is sum of squares residual error
#SST sum of squares total 
```

```python
def SquaredError(y_original,y_line):
    return sum((y_original-y_line)**2)
```

```python
def Coefficient_r_value(y,regression_line):
    mean_array=[]
    for i in y:
        mean_array.append(mean(y))
        
    SSE=SquaredError(y,regression_line)
    SST=SquaredError(y,mean_array)
    return 1-(SSE/SST)
```

```python
rsquared=Coefficient_r_value(y,regression_line)
```
